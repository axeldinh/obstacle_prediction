```python
from load_data import load_data
import numpy as np
from utils import save_submission
```

```python
test_data, _ = load_data('test')

filtered_data = []

for d in test_data:

    nan_index = np.where(d==-1)[0]
    min_nan = np.min(nan_index)

    filtered_data.append(d[:min_nan])
```

```python
predictions = np.array(
    [0 if len(d) < 36 else 1 for d in filtered_data]
).astype(int)
```

```python
import os, shutil
os.makedirs('submission', exist_ok=True)
save_submission(predictions, 'submission/submission.csv')

# Get the current filename

shutil.copyfile('size_filter.ipynb', 'submission/original_notebook.ipynb')
shutil.make_archive('submission', 'zip', 'submission')
```

```python
import matplotlib.pyplot as plt
for d in filtered_data:
    
    d = d[np.argsort(d[:, 1])]

    plt.plot(d[:, 1], d[:, 0])
    plt.plot(d[:, 1], d[:, 2])
    plt.plot(d[:, 1], d[:, 3])

    plt.show()
```
